```python
import zipfile
import os

zip_path = "Machine-Learning/Agatha_Christie_Corpus.zip"
extract_path = "Machine-Learning/extracted_files"

# Create a directory to extract if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List extracted files
!ls Machine-Learning/extracted_files
```

```python
file_path = "Machine-Learning/extracted_files/Poirot Investigates_pg61262.txt"

with open(file_path, "r", encoding="utf-8") as file:
    content = file.read()
    print(content[:10000])
```

```python
import pandas as pd

# Lists to store filenames and text contents
filenames = []
texts = []

# Loop through all .txt files in the directory
for filename in os.listdir(extract_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(extract_path, filename)
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            filenames.append(filename)  # Store filename
            texts.append(file.read())   # Store file content

# Create a DataFrame
df = pd.DataFrame({"Filename": filenames, "Content": texts})
```
