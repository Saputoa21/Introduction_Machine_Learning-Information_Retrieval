## 2. Creating the Dataframe from Texts

```python
import pandas as pd

# Lists to store filenames and text contents
filenames = []
texts = []

# Loop through all .txt files in the directory
for filename in os.listdir(extract_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(extract_path, filename)
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            filenames.append(filename)
            texts.append(file.read())

# Create a DataFrame
df = pd.DataFrame({"file_name": filenames, "content": texts})
```

## 3. Cleaning and Preprocessing on One Book

### Removing metadata of Gutenberg Project

```python
book_1 = df['content'][0]
```

```python
# Remove empty lines and join all lines into one
cleaned_book_1 = "\n".join([line.strip() for line in book_1.split("\n") if line.strip()])

print(cleaned_book_1[:1000])
```

```python
# Split lines and find where the actual content starts and ends
lines = cleaned_book_1.split("\n")
start_idx = 0
end_idx = 0

for i, line in enumerate(lines):
  if "*** START OF THE PROJECT GUTENBERG EBOOK" in line:
    start_idx = i + 1
  if "*** END OF THE PROJECT GUTENBERG EBOOK" in line:
    end_idx = i
    break

book_1_main_content = " ".join(lines[start_idx:end_idx])

print(book_1_main_content[:1000])
```

### Tokenization and Lemmatisation

```python
!python -m spacy download en_core_web_trf
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
processed_book_1 = nlp(book_1_main_content)
```

```python
# Tokenisation

tokens = []

for token in processed_book_1:
  tokens.append(token.text)

unique_tokens = set(tokens)

print(f'Number of tokens in the preproccessed text: {len(tokens)}')
print(f'Number of unique tokens  in the preproccessed text: {len(unique_tokens)}\n')

print(tokens[:100])

# Lemmatisation

lemmatised_tokens = []

for token in processed_book_1:
  lemmatised_tokens.append(token.lemma_)

print(lemmatised_tokens[:100])
```

### Removing stopwords and punctuation

```python
import nltk
import string
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)
```

```python
print(stop_words)
print(punctuation)
```

```python
# Removing extra punctutation and stopwords

clean_tokens = []
for token in tokens:
  if token in punctuation or token in stop_words:
    continue
  else:
    clean_tokens.append(token)

print(clean_tokens[:100])

clean_lemmatised_tokens = []
for token in lemmatised_tokens:
  if token in punctuation or token in stop_words:
    continue
  else:
    clean_lemmatised_tokens.append(token)

print(clean_lemmatised_tokens[:100])

print(len(clean_tokens))
print(len(clean_lemmatised_tokens))
```
