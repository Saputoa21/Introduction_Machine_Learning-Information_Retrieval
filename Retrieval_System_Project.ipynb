## 2. Creating the Dataframe from Texts

```python
import pandas as pd

# Lists to store filenames and text contents
filenames = []
texts = []

# Loop through all .txt files in the directory
for filename in os.listdir(extract_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(extract_path, filename)
        with open(file_path, "r", encoding="utf-8", errors="ignore") as file:
            filenames.append(filename)  # Store filename
            texts.append(file.read())   # Store file content

# Create a DataFrame
df = pd.DataFrame({"Filename": filenames, "Content": texts})
```

## 3. Cleaning and Preprocessing on One Book

```python
book_1 = df['Content'][0]
```

```python
# Remove empty lines
cleaned_book_1 = "\n".join([line.strip() for line in book_1.split("\n") if line.strip()])
```

```python
# Split lines and find the actual content start and ends
lines = cleaned_book_1.split("\n")
start_idx = 0
end_idx = 0

for i, line in enumerate(lines):
  if "*** START OF THE PROJECT GUTENBERG EBOOK" in line:
    start_idx = i + 1
  if "*** END OF THE PROJECT GUTENBERG EBOOK" in line:
    end_idx = i
    break

# Extract the main content
book_1_main_content = "\n".join(lines[start_idx:end_idx])

# Print the cleaned content
print(book_1_main_content)
```

```python
processed_book_1 = nlp(book_1_main_content)
```

```python
# Tokenisation

tokens = []

for token in processed_book_1:
  tokens.append(token.text)

unique_tokens = set(tokens)

print(f'Number of tokens in the preproccessed text: {len(tokens)}')
print(f'Number of unique tokens  in the preproccessed text: {len(unique_tokens)}')

print(tokens[:10]) # there are '\n' signs that need to be removed

# Lematisation

lemmatised_tokens = []

for token in processed_book_1:
  lemmatised_tokens.append(token.lemma_)

print(lemmatised_tokens[:10])

# Removing '\n's

new_tokens = []
for token in tokens:
  if token != '\n':
    new_tokens.append(token)
  continue

print(new_tokens[:10])

new_lemmatised_tokens = []
for token in lemmatised_tokens:
  if token != '\n':
    new_lemmatised_tokens.append(token)
  continue

print(new_lemmatised_tokens[:10])
```
